```python
# https://spacy.io/
!pip3 install spacy
# spacy 예측기반 모델 hence download weight set
# 사전훈련된 nlp 모델 다운로드 
# https://spacy.io/models/en#en_core_web_sm
!python -m spacy download en_core_web_sm
```

```python
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('merge_entities')

print(CORPUS[1])  # original text
```

```python
t_CORPUS = [
    [token.lemma_ for token in nlp(str(c)) if token.pos_ != "PUNCT"]
    for c in CORPUS
]
print(t_CORPUS[1])  # tokenized text
```

```python
model_gen.build_vocab(t_CORPUS, update=True)  # Update the vocab list (further training)

model_gen.train(t_CORPUS, total_examples=len(t_CORPUS), epochs=model_gen.epochs)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

print(cosine_similarity(model_gen.wv['Corleone'].reshape(1,-1), model_gen.wv['mafia'].reshape(1,-1)))
print(cosine_similarity(model_gen.wv['shawshank'].reshape(1,-1), model_gen.wv['music'].reshape(1,-1)))
print(cosine_similarity(model_gen.wv['batman'].reshape(1,-1), model_gen.wv['action'].reshape(1,-1)))
```

# TFIDF matrix

```python
# noun_words = [
#            [(token.lemma_, token.pos_) for token in nlp(str(c)) if token.pos_ == "NOUN"]
#            for c in CORPUS[:10]
# ]
# tfidf_CORPUS = [" ".join(c) for c in noun_words]
# print(CORPUS[1])
# print(tfidf_CORPUS[1])  # only noun words can be a keyword
```

```python
tfidf_CORPUS = [
           [token.lemma_ for token in nlp(str(c)) if token.pos_ not in ['PUNCT', 'PART']]  # PROPN
           for c in CORPUS
]
# tfidf_CORPUS = t_CORPUS

tfidf_CORPUS[:5]
```

```python
tfidf_CORPUS = [" ".join(c) for c in tfidf_CORPUS]
tfidf_CORPUS[:3]
```

```python
len(tfidf_CORPUS)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()  # stop_words?  stop_words='english'
matrix = vectorizer.fit_transform(tfidf_CORPUS)

words = vectorizer.get_feature_names()
dtm_tfidf = matrix.toarray()

print(len(words), words[1000:1010])
print(dtm_tfidf[3][1000:1010])
```

```python
dtm_tfidf.shape
```

```python
top_tfidf_scores = np.argsort(dtm_tfidf[1])[-10:]  # top 10 overview keywords
[words[w] for w in top_tfidf_scores][::-1]
```

```python
TOP_N = 10
movie_keywords = []
for movie in dtm_tfidf:
    top_tfidf_scores = np.argsort(movie)[-TOP_N:]  # top N overview scores
    top_tfidf_scores = [(words[w], movie[w]) for w in top_tfidf_scores][::-1]
    movie_keywords.append(top_tfidf_scores)
    
list(zip(titles[:5], movie_keywords[:5]))
```

# Keyword TEST

```python
[[words[0] for words in movie] for movie in movie_keywords]
```

```python
# userKeywords = ['gun', 'suitcase', 'cigarette', 'sunglasses', 'car']
userKeywords = ['robe', 'book', 'newspaper', 'wand', 'cake', 'broom']
```

```python
score_list = []
for movie in movie_keywords:
    score = 0
    for word in movie:
        for userKeyword in userKeywords:
#             print(word[0], userKeyword)
            score += float(cosine_similarity(model_gen.wv[word[0]].reshape(1,-1), model_gen.wv[userKeyword].reshape(1,-1)))  # * word[1]
    score_list.append(score)
    
list(zip(titles, score_list))
# score_list
```

```python
result_df = pd.DataFrame(score_list, index=titles, columns=['score'])
result_df.sort_values('score', ascending=False)[:10]
```
